```python
from scipy.optimize import curve_fit
```

```python
def f(x, a, b, c):
    x = np.clip(x, 0, 10000)
    return a * (x)**b + x*c
```

```python
def get_fit(f, x, y):
    ff = curve_fit(f, x, y, p0=[1,0.5, 0], maxfev = int(1e5))
    print(ff)
    return ff[0]
```
